```python
import pandas as pd

data = pd.read_csv('EmployeeAttrition.csv')
```

```python
print(data.head())
```

```python
#Check the head of the DataFrame.
print(data.head())
print(data.count)
```

```python
#Q1: Find the number of entries/rows and columns in the data.
 
x=data.shape[0]
y=data.shape[1]
print("1)   No of Rows and Columns: ")
print(x,y)
#Q2: What is the average Monthly Income?

m=data["MonthlyIncome"].mean()
print("2) The Mean Monthly salary is ")
print(m)

#Q3: What is the highest amount of HourlyRate ?
x=data["HourlyRate"].max()
print("3) Highest amount of HourlyRate is")
print(x)

#Q4: What is the Department, JobRole, MaritalStatus and OverTime of EmployeeNumber 10?
df = pd.DataFrame(data, columns = ['Department','JobRole','MaritalStatus','OverTime','EmployeeNumber'])
x=df[df['EmployeeNumber']==10]
print("4) The Department, JobRole, MaritalStatus and OverTime of EmployeeNumber 10")
print(x)

#Q5: What is the Employee ID of highest MonthlyIncome paid employee?
df = pd.DataFrame(data, columns = ['EmployeeNumber','MonthlyIncome'])
x=df[df['MonthlyIncome']==df['MonthlyIncome'].max()]
print("5) The Employee ID of highest MonthlyIncome paid employee")
print(x)

#Q6: What is the average(mean) DailyRate group by Age for all Employees whose age is greater than 58. (hint: use groupby function)
df= pd.DataFrame(data, columns = ['Age','DailyRate'])
df=df[df['Age']>58]
x=df.groupby('DailyRate').mean()
print("6) The average(mean) DailyRate group by Age for all Employees whose age is greater than 58")
print(x)

#Q7: How many unique EducationField are there?
df=data['EducationField'].unique()
x=len(df)
print("7) No of unique EducationField:")
print(x)

#Q8: What are the top 5 most common JobRole?
df = pd.DataFrame(data,columns=['JobRole','EmployeeNumber'])
df1=df.groupby(['JobRole'])['EmployeeNumber'].agg(['count'])
df2=df1.sort_values(by='count', ascending=False).head(5)
print("8) Top 5 most common JobRole:")
print(df2)

#Q9: How many JobRoles represented by less than 100 employees?
df = pd.DataFrame(data,columns=['JobRole','EmployeeNumber'])
df1=df.groupby(['JobRole'])['EmployeeNumber'].agg(['count'])
df2=df1[df1['count']<100]
print("9) JobRoles represented by less than 100 employees")
print(df2)

#Q10: What is the correlation between Education and JobSatisfaction?
df=data[['Education','JobSatisfaction']]
correlation = df.corr(method='pearson')
print(correlation)
print("10) Correlation is -0.011296")
```
